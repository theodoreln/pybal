## Extract data

```python
unit_results = Data.get_symbol('GKDERATE')
```

```python
unit_results
```

```python
# Remove scenario column for grouping
unit_results = unit_results.drop(columns=['Scenario', 'SSS', 'TTT'])

# I want the average over S and T columns of values for all combinations of Scenario, AAA and GGG
avg_values = unit_results.groupby(['AAA', 'GGG']).mean()

# Delete all index with no values (all NaN)
avg_values = avg_values.dropna(how='all')
```

```python
# Export average values to CSV
avg_values.to_csv("average_gkderate_values.csv")
```
